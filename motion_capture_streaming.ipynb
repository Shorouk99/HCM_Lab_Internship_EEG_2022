```python
import asyncio
import qtm
from pylsl import StreamInfo, StreamOutlet
```

```python
channel_names = ['x','y','z']
info = StreamInfo('QTM', 'Motion Capture', 3, 300, 'float32', '')
outlet = StreamOutlet(info)
# chns = info.desc().append_child("channels")
# for label in channel_names:
#     ch = chns.append_child("channel")
#     ch.append_child_value("label", label)
```

```python
def on_packet(packet):
    header, markers = packet.get_3d_markers()
    
    for marker in markers:
        marker = str(marker)[19:-1].split(',')
        key = [string.split('=')[0] for string in marker]
        value = [float(string.split('=')[1]) for string in marker]
        dictionary = dict(zip(key,value))
        
        outlet.push_chunk(list(dictionary.values()))
        print (dictionary)
```

```python
async def setup():
    """ Main function """
    connection = await qtm.connect("127.0.0.1")
    if connection is None:
        return

    await connection.stream_frames(components=["3d"], on_packet=on_packet)
```
