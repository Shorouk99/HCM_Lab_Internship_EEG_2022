# LSL Latency Measurement Using RTT

Samples from another device were generated and timestamped using the other device's local clock. This script recieves the samples as well as their timestamps, and send the timestamps back to the original sending-device to measure the Round Trip Time (RTT) and calculate the latency

```python
from pylsl import StreamInlet, resolve_stream, StreamInfo, StreamOutlet
```

At first, we will search for the stream which we want to measure its latency. While creating this notebook, we used a motion capture stream generated by QTM, which we set its type to "Motion Capture" at time of creating the stream. Replace it with the type of the stream you wish to measure its latency.

```python
print("looking for Motion Capture stream...")
streams = resolve_stream('type', 'Motion Capture')
inlet = StreamInlet(streams[0])
```

Then, we will create another stream to output the timestamps to the network. This streaming will continue infinitly untill the kernel is restarted. The next step is to use the second script to recieve the timestamps on the original sending device and perform the required analysis

```python
info = StreamInfo('RTT', 'Latency Analysis', 1, 300, 'float32', 'multi5')
outlet = StreamOutlet(info)
```

```python
while True:
    sample, timestamp = inlet.pull_sample()
    outlet.push_sample([timestamp])
    print(timestamp, sample)
```
